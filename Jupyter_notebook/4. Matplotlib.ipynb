```python
import numpy as np
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
#先来个简单的曲线试试
x=np.arange(0.1,4,.01)
plt.figure(figsize=(20,10))                              #设置图片尺寸（比例）
plt.plot(x,np.sin(x*50)/x,linewidth=3)                   #函数为y=sin(50x)/x
plt.plot(x,np.zeros(len(x)),'g--',linewidth=2)           #在0处画一条虚线样式的水平线，默认蓝色实线为'b-'
plt.title('Cureve $y=\sin(50x)/x$',fontsize=40,y=1.04)   #标题，这里书上有个小bug多了个};y=1.04定义了标题的竖直空间
plt.xlabel('$x$',fontsize=30)
plt.ylabel('$y=\sin(50x)/x$',fontsize=30)
plt.grid(True)                                           #背景网格线：yes
plt.savefig('mplsin.png')                                #存入文件
```

```python
#如果要在同一张背景里画多条曲线的话
x=np.arange(0,2,.05)
plt.figure(figsize=(10,5))
plt.plot(x,np.cos(x),'r:',x,np.cos(x**2),'b^',linewidth=10,markersize=15)        #r:和b^表示线型/点型
plt.ylim((-1.5,1.5))
plt.title('2 curves in one figure',fontsize=30)
```
